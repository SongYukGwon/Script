    >>> c = c + 5

#### 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
    >>> print(c > 10)
    True
    >>> print(d > 10)
    False
    >>> print(c < 10)
    False
    >>> print(d < 10)
    True
    >>> print(c == 17)
    True
    >>> print(d != 9)
    False

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    def __str__(self):
        return f"[Count (step: {self.step})] {self.count}"
    def incr(self):
        self.count += self.step
    def __call__(self):
        self.incr()
    def __add__(self, other):
        tmp = Counter(self.count, self.step)
        tmp.count += other
        return tmp
    def __sub__(self, other):
        tmp = Counter(self.count, self.step)
        tmp.count -= other
        return tmp
    def __lt__(self, other):
        return self.count < other
    def __gt__(self, other):
        return self.count > other
    def __eq__(self, other):
        return self.count == other
    
c = Counter(10)
d = Counter(10, 2)
print(c)
print(d)
```

```python
c.incr()
d.incr()
print(c)
print(d)
```

### 4.다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

#### 1. __init__()
    __init__메소드는 인스턴스가 생성될때 자동으로 호출되는 함수로서 초기화 함수라고 생각하면 된다. 위에서는 __init__에 매개변수로 l이라는 리스트를 받아 list를 상속받은 자신에게 추가해주고 가지고 있는 eliminate_duplicate를 실행하는것을 보여준다.
    
#### 2. __str__()
    __str__메소드는 출력 연산자 오버로딩으로 print(s)같이 인스턴스 변수명을 출력하고자하면 실행되는 함수이다. 위에서는 인스턴스가 가지고 있는 원소들을 형태에 맞춰 출력하도록 오버로딩하였다.
    
#### 3.eliminate_duplicate()
    eliminate_duplicate함수는 기능부터 설명하자면 __init__을 통해 들어온 원소들중 중복되어 있는 것을 없애는 함수이다. 그런데 여기서 봐야할것은 데코레이터(@)로 @staticmethod이 붙어있는데 이것은 정적메소드라는 선언으로 객체를 만들지 않고도 사용할 수 있는 메소드라는 뜻이다. 그래서 다른 함수들과 비교하였을때는 self를 매개변수로 사용하지 않는다는 특이점이 있습니다. 일반적으로 정적 메서드는 메서드의 실행이 외부 상태에 영향을 끼치지 않는 순수 함수(pure function)을 만들때 사용합니다.

### 5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.
#### 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
#### 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

    >>> u = s | t
    >>> print(u)
    MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}

#### 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

    >>> u = s & t
    >>> print(u)
    MySet: {2, 3}

#### 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

    >>> s = MySet([1, 2, 3])
    >>> t = MySet([3, 4, 5])
    >>> u = s - t
    >>> print(u)
    MySet: {1, 2}

```python
class MySet(list):
    def __init__(self, l=[]):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __and__(self, other): #&연산자 오버로딩
        tmp = MySet()
        for i in other:
            if i in self: #겹치는게 있으면 tmp에 추가 후 return
                tmp.append(i)
        return tmp

    def __or__(self, other): #|연산자 오버로딩
        tmp = MySet()
        for i in self: #self에 있는것을 추가 후
            tmp.append(i)
        for i in other:
            if i not in tmp: #추가한것에 other에 있는것이 없으면 추가 후 리턴
                tmp.append(i)
        return tmp
    
    def __sub__(self, other): # - 연산자 오버로딩
        for i in other:
            if i in self: #self에 other의 원소가 있으면 삭제 후 리턴
                self.remove(i)
        return self

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    print()
    u = s | t
    print("s or t -> ",u)
    print()
    u = s & t
    print("s and t -> ", u)
    print()
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print("s - t -> ", u)
```

#### 1. __or__()오버로딩으로 객체끼리 |연산자 사용시 두 객체의 합을 구하여 return 하도록 만들었다.
#### 2. __and__()오버로딩으로 객체끼리 &연산자 사용시 두객체에서 겹치는것만들 return 하도록 만들었다.
#### 3. __sub__()오버로딩으로 객체끼리 -연산자 사용시 앞에있는 객체에서 뒤에있는 객체와 겹치는것을 삭제하고 return하도록 만들었다.

```python
import requests
import string

class SearchEngine():
    #초기화
    def __init__(self, *args):
        self.urls = []
        self.word_dict = {}
        for i in args:
            self.urls.append(i)
    
    #url 추가 메소드
    def addUrl(self,url):
        self.urls.append(url)
        
    #url 삭제 메소드
    def removeUrl(self,url):
        if url in self.urls:
            self.urls.remove(url)
            
    #url정보 반환 메소드
    def listUrls(self):
        for i in self.urls:
            print(i)
    
    #빈도수 얻는 메소드
    def getWordsFrequency(self):
        if len(self.urls) == 0:
            return {}
        
        for i in self.urls:
            dic = self.getWord(i)
            for key, value in dic.items():
                if key in self.word_dict.keys():
                    self.word_dict[key] += value
                else:
                    self.word_dict[key] = value
        return self.word_dict
    
    #최대빈도수 단어 반환 메소드
    def getMaxFreqencyWords(self):
        if len(self.urls) == 0:
            return None
        self.word_dict = dict(sorted(self.word_dict.items(), key = lambda x: x[1], reverse = True))
        max_dict = {}
        max_fre = -1
        for key, value in self.word_dict.items():
            if max_fre == -1:
                max_fre = value
            if value == max_fre:
                max_dict[key] = value
            elif value < max_fre:
                break;
        return max_dict
    
    # 검색메소드
    def searchUrlByWord(self, word):
        score = {}
        total = 0;
        for url in self.urls:
            data = self.getWord(url)
            point = self.similiar(word, data)
            score[url] = point
            total += point
        #total이 0일경우 에러방지를 위해 total을 1로 설정
        if total == 0:
            total = 1
        #결과를 point기준으로 내림차순 정렬
        score_sorted= dict(sorted(score.items(), key = lambda x: x[1], reverse = True))
        max_list = []
        max_fre = -1
        for key, value in score_sorted.items():
            if max_fre == -1:
                max_fre = value
            if value == max_fre:
                max_list.append(key)
            elif value < max_fre:
                break;
        return max_list
    
    #유사도 검사메소드
    def similiar(self, inp1, inp2):
        total_point = 0
        check_point = 0
        for key, value in inp2.items():
            total_point += value
            if key in inp1:
                check_point += value
        if total_point == 0:
            total_point= 1
        return check_point/total_point
    
    # 단어 빈도수얻는 메소드
    def getWord(self, url):
        req = requests.get(url)
        source = req.text

        #단어추출
        check = 0
        check_slash =0
        total = ""
        for i in range(len(source)):
            if (source[i:i+7] == '<script' or source[i:i+6] == '<style') and check_slash == 0:
                    check_slash = 1
            elif (source[i:i+9] == '</script>' or source[i:i+8] == '</style>') and check_slash == 1:
                    check_slash = 0

            if check == 0:
                if source[i] =="<":
                    check =1
                else:
                    if check_slash ==0:
                        total += source[i]
            else:
                if source[i] == ">":
                    check =0
        li = total.split()
        result = []
        for i in li:
            for punctuation in string.punctuation:
                i = i.replace(punctuation, "")

            #불용어면 추가하지않고 건너뜀
            if i.lower() in bull_word:
                continue
            result.append(i.lower())
        result = list(filter(None, result))

        word_dic = {}
        for i in result:
            if i in word_dic:
                word_dic[i] += 1
            else:
                word_dic[i]=1
        return word_dic
```

### 2. 위 1번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

    리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
    getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.

    reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
    reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
        w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')

        w4.getWordsFrequency()
        [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]

        w4.getWordsFrequency(reverse=true)
        [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    def __init__(self, *args):
        super(SearchEngineWithOrderedWebWords, self).__init__(*args)
        
    def getWordsFrequency(self, reverse = False):
        if reverse == False:
            reverse = True
        elif reverse == True:
            reverse = False
        if len(self.urls) == 0:
            return {}
        
        for i in self.urls:
            dic = self.getWord(i)
            for key, value in dic.items():
                if key in self.word_dict.keys():
                    self.word_dict[key] += value
                else:
                    self.word_dict[key] = value
            word_list_sorted = list(sorted(self.word_dict.items(), key = lambda x:x[1], reverse = reverse))
        return word_list_sorted
    
    def __iter__(self):
        self.n = 0;
        self.word_sorted = list(sorted(self.word_dict.items(), key = lambda x:x[1], reverse = True))
        return self
    
    def __next__(self):
        if self.n == len(self.word_sorted):
            raise StopIteration
        self.n +=1
        return self.word_sorted[self.n-1]
```

__iter__() 와 __next__()를 사용하여 for문에 해당 객체를 넣을 수 있도록 만들었다.
__iter__()에는 예시처럼 내림차순으로 출력할 수 있도록 내림차순으로 정렬한 리스트를 만들어 주었고, 위치를 알 수 있도록 n을 선언해준뒤 return self를 해주었다.
__next__()에는 불러올때마다 n을 하나씩 더해준뒤 n-1번째에 위치해있는 것을 출력하도록 하였다. 만약 n이 가지고 있는 단어 리스트의 길이만큼 된다면 StopIteration을 raise해주어 끝내도록 해주었다.
