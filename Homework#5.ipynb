#### ìš”êµ¬ì‚¬í•­ 6. ë‹¤ìŒê³¼ ê°™ì€ ê´€ê³„ì—°ì‚° (>, <, ==)ì´ ìˆ˜í–‰ë  ìˆ˜ ìˆë„ë¡ Counter í´ë˜ìŠ¤ ë‚´ì— ê´€ë ¨ ë©”ì†Œë“œë¥¼ ì¶”ê°€í•˜ì‹œì˜¤.
    >>> print(c > 10)

    >>> print(c < 10)

    >>> print(d < 10)

    >>> print(c == 17)
    True
    >>> print(d != 9)
    False

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    def __str__(self):
        return f"[Count (step: {self.step})] {self.count}"
    def incr(self):
        self.count += self.step
    def __call__(self):
        self.incr()
    def __add__(self, other):
        tmp = Counter(self.count, self.step)
        tmp.count += other
        return tmp
    def __sub__(self, other):
        tmp = Counter(self.count, self.step)
        tmp.count -= other
        return tmp
    def __lt__(self, other):
        return self.count < other
    def __gt__(self, other):
        return self.count > other
    def __eq__(self, other):
        return self.count == other
    
c = Counter(10)
d = Counter(10, 2)
print(c)
print(d)
```

### 4.ë‹¤ìŒì€ ë‚´ì¥ ìë£Œí˜• listë¥¼ ì„œë¸Œí´ë˜ì‹±í•˜ì—¬ ë§Œë“  MySet í´ë˜ìŠ¤ ì •ì˜ ë‚´ìš©ì´ë‹¤. ë‹¤ìŒ í´ë˜ìŠ¤ ì •ì˜ì—ì„œ __init__(), __str()__(), elimicate_duplicate()ì˜ ì„¸ ê°œì˜ ë©”ì†Œë“œ ì½”ë“œ ë‚´ìš©ì„ ìì‹ ì´ ë‹¤ë¥¸ ì‚¬ëŒì—ê²Œ ê°€ë¥´ì¹œë‹¤ê³  ìƒê°í•˜ë©° ì„¤ëª…í•´ë³´ì‹œì˜¤.

MySetì€ ì§‘í•©(Set) ìë£Œí˜•ì„ ì •ì˜í•˜ë ¤ëŠ” ì˜ë„í•˜ì— ë§Œë“¤ì–´ì§„ í´ë˜ìŠ¤ì´ë‹¤.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

    eliminate_duplicateí•¨ìˆ˜ëŠ” ê¸°ëŠ¥ë¶€í„° ì„¤ëª…í•˜ìë©´ __init__ì„ í†µí•´ ë“¤ì–´ì˜¨ ì›ì†Œë“¤ì¤‘ ì¤‘ë³µë˜ì–´ ìˆëŠ” ê²ƒì„ ì—†ì• ëŠ” í•¨ìˆ˜ì´ë‹¤. ê·¸ëŸ°ë° ì—¬ê¸°ì„œ ë´ì•¼í• ê²ƒì€ ë°ì½”ë ˆì´í„°(@)ë¡œ @staticmethodì´ ë¶™ì–´ìˆëŠ”ë° ì´ê²ƒì€ ì •ì ë©”ì†Œë“œë¼ëŠ” ì„ ì–¸ìœ¼ë¡œ ê°ì²´ë¥¼ ë§Œë“¤ì§€ ì•Šê³ ë„ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” ë©”ì†Œë“œë¼ëŠ” ëœ»ì´ë‹¤. ê·¸ë˜ì„œ ë‹¤ë¥¸ í•¨ìˆ˜ë“¤ê³¼ ë¹„êµí•˜ì˜€ì„ë•ŒëŠ” selfë¥¼ ë§¤ê°œë³€ìˆ˜ë¡œ ì‚¬ìš©í•˜ì§€ ì•ŠëŠ”ë‹¤ëŠ” íŠ¹ì´ì ì´ ìˆìŠµë‹ˆë‹¤. ì¼ë°˜ì ìœ¼ë¡œ ì •ì  ë©”ì„œë“œëŠ” ë©”ì„œë“œì˜ ì‹¤í–‰ì´ ì™¸ë¶€ ìƒíƒœì— ì˜í–¥ì„ ë¼ì¹˜ì§€ ì•ŠëŠ” ìˆœìˆ˜ í•¨ìˆ˜(pure function)ì„ ë§Œë“¤ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.

### 5. 4ë²ˆ ë¬¸ì œì— ì •ì˜ëœ MySet í´ë˜ìŠ¤ì— ë©”ì†Œë“œë¥¼ ì¶”ê°€í•˜ì—¬ ë‹¤ìŒ ê° ìš”êµ¬ì‚¬í•­ ëª¨ë‘ë¥¼ ë§Œì¡±ì‹œí‚¤ëŠ” ì½”ë”©ì„ ì œì‹œí•˜ì‹œì˜¤.
#### ì •ë‹µì„ ê° ìš”êµ¬ì‚¬í•­ë³„ë¡œ ì…ë ¥í•  í•„ìš” ì—†ì´ ìš”êµ¬ì‚¬í•­ 3ê°œ ì „ì²´ì— ëŒ€í•´ 1ê°œì˜ MySet í´ë˜ìŠ¤ ì •ì˜ ì½”ë“œë¥¼ ì œì‹œí•˜ê³  ì£¼ì„ë“±ìœ¼ë¡œ í•´ë‹¹ ì½”ë“œë¥¼ ì„¤ëª…í•œë‹¤.
#### ìš”êµ¬ì‚¬í•­ 1. | ì—°ì‚°ìœ¼ë¡œ ë‘ ì§‘í•©ì˜ í•©ì§‘í•©ì„ ë°˜í™˜í•œë‹¤.

#### ìš”êµ¬ì‚¬í•­ 2. & ì—°ì‚°ìœ¼ë¡œ ë‘ ì§‘í•©ì˜ êµì§‘í•©ì„ ë°˜í™˜í•œë‹¤.

#### ìš”êµ¬ì‚¬í•­ 3. - ì—°ì‚°ìœ¼ë¡œ ë‘ ì§‘í•©ì˜ ì°¨ì§‘í•©ì„ ë°˜í™˜í•œë‹¤.

    >>> s = MySet([1, 2, 3])
    >>> t = MySet([3, 4, 5])

```python
class MySet(list):
    def __init__(self, l=[]):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __and__(self, other): #&ì—°ì‚°ì ì˜¤ë²„ë¡œë”©
        tmp = MySet()
        for i in other:
            if i in self: #ê²¹ì¹˜ëŠ”ê²Œ ìˆìœ¼ë©´ tmpì— ì¶”ê°€ í›„ return
                tmp.append(i)
        return tmp

    def __or__(self, other): #|ì—°ì‚°ì ì˜¤ë²„ë¡œë”©
        tmp = MySet()
        for i in self: #selfì— ìˆëŠ”ê²ƒì„ ì¶”ê°€ í›„
            tmp.append(i)
        for i in other:
            if i not in tmp: #ì¶”ê°€í•œê²ƒì— otherì— ìˆëŠ”ê²ƒì´ ì—†ìœ¼ë©´ ì¶”ê°€ í›„ ë¦¬í„´
                tmp.append(i)
        return tmp
    
    def __sub__(self, other): # - ì—°ì‚°ì ì˜¤ë²„ë¡œë”©
        for i in other:
            if i in self: #selfì— otherì˜ ì›ì†Œê°€ ìˆìœ¼ë©´ ì‚­ì œ í›„ ë¦¬í„´
                self.remove(i)
        return self

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    print()
    u = s | t
    print("s or t -> ",u)
    print()
    u = s & t
    print("s and t -> ", u)
    print()
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print("s - t -> ", u)
```

```python
import requests
import string

class SearchEngine():
    #ì´ˆê¸°í™”
    def __init__(self, *args):
        self.urls = []
        self.word_dict = {}
        for i in args:
            self.urls.append(i)
    
    #url ì¶”ê°€ ë©”ì†Œë“œ
    def addUrl(self,url):
        self.urls.append(url)
        
    #url ì‚­ì œ ë©”ì†Œë“œ
    def removeUrl(self,url):
        if url in self.urls:
            self.urls.remove(url)
            
    #urlì •ë³´ ë°˜í™˜ ë©”ì†Œë“œ
    def listUrls(self):
        for i in self.urls:
            print(i)
    
    #ë¹ˆë„ìˆ˜ ì–»ëŠ” ë©”ì†Œë“œ
    def getWordsFrequency(self):
        if len(self.urls) == 0:
            return {}
        
        for i in self.urls:
            dic = self.getWord(i)
            for key, value in dic.items():
                if key in self.word_dict.keys():
                    self.word_dict[key] += value
                else:
                    self.word_dict[key] = value
        return self.word_dict
    
    #ìµœëŒ€ë¹ˆë„ìˆ˜ ë‹¨ì–´ ë°˜í™˜ ë©”ì†Œë“œ
    def getMaxFreqencyWords(self):
        if len(self.urls) == 0:
            return None
        self.word_dict = dict(sorted(self.word_dict.items(), key = lambda x: x[1], reverse = True))
        max_dict = {}
        max_fre = -1
        for key, value in self.word_dict.items():
            if max_fre == -1:
                max_fre = value
            if value == max_fre:
                max_dict[key] = value
            elif value < max_fre:
                break;
        return max_dict
    
    # ê²€ìƒ‰ë©”ì†Œë“œ
    def searchUrlByWord(self, word):
        score = {}
        total = 0;
        for url in self.urls:
            data = self.getWord(url)
            point = self.similiar(word, data)
            score[url] = point
            total += point
        #totalì´ 0ì¼ê²½ìš° ì—ëŸ¬ë°©ì§€ë¥¼ ìœ„í•´ totalì„ 1ë¡œ ì„¤ì •
        if total == 0:
            total = 1
        #ê²°ê³¼ë¥¼ pointê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
        score_sorted= dict(sorted(score.items(), key = lambda x: x[1], reverse = True))
        max_list = []
        max_fre = -1
        for key, value in score_sorted.items():
            if max_fre == -1:
                max_fre = value
            if value == max_fre:
                max_list.append(key)
            elif value < max_fre:
                break;
        return max_list
    
    #ìœ ì‚¬ë„ ê²€ì‚¬ë©”ì†Œë“œ
    def similiar(self, inp1, inp2):
        total_point = 0
        check_point = 0
        for key, value in inp2.items():
            total_point += value
            if key in inp1:
                check_point += value
        if total_point == 0:
            total_point= 1
        return check_point/total_point
    
    # ë‹¨ì–´ ë¹ˆë„ìˆ˜ì–»ëŠ” ë©”ì†Œë“œ
    def getWord(self, url):
        req = requests.get(url)
        source = req.text

        #ë‹¨ì–´ì¶”ì¶œ
        check = 0
        check_slash =0
        total = ""
        for i in range(len(source)):
            if (source[i:i+7] == '<script' or source[i:i+6] == '<style') and check_slash == 0:
                    check_slash = 1
            elif (source[i:i+9] == '</script>' or source[i:i+8] == '</style>') and check_slash == 1:
                    check_slash = 0

            if check == 0:
                if source[i] =="<":
                    check =1
                else:
                    if check_slash ==0:
                        total += source[i]
            else:
                if source[i] == ">":
                    check =0
        li = total.split()
        result = []
        for i in li:
            for punctuation in string.punctuation:
                i = i.replace(punctuation, "")

            #ë¶ˆìš©ì–´ë©´ ì¶”ê°€í•˜ì§€ì•Šê³  ê±´ë„ˆëœ€
            if i.lower() in bull_word:
                continue
            result.append(i.lower())
        result = list(filter(None, result))

        word_dic = {}
        for i in result:
            if i in word_dic:
                word_dic[i] += 1
            else:
                word_dic[i]=1
        return word_dic
```

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    def __init__(self, *args):
        super(SearchEngineWithOrderedWebWords, self).__init__(*args)
        
    def getWordsFrequency(self, reverse = False):
        if reverse == False:
            reverse = True
        elif reverse == True:
            reverse = False
        if len(self.urls) == 0:
            return {}
        
        for i in self.urls:
            dic = self.getWord(i)
            for key, value in dic.items():
                if key in self.word_dict.keys():
                    self.word_dict[key] += value
                else:
                    self.word_dict[key] = value
            word_list_sorted = list(sorted(self.word_dict.items(), key = lambda x:x[1], reverse = reverse))
        return word_list_sorted
    
    def __iter__(self):
        self.n = 0;
        self.word_sorted = list(sorted(self.word_dict.items(), key = lambda x:x[1], reverse = True))
        return self
    
    def __next__(self):
        if self.n == len(self.word_sorted):
            raise StopIteration
        self.n +=1
        return self.word_sorted[self.n-1]
```
