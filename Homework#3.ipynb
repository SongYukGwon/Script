### 5.다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.

1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

```python
def addall(li):
    result =0
    for i in li:
        result+= i
    return result

print(addall([1]))
print(addall([1,2,3,4,5]))
```

#### addall에 list인자를 받은다음 result로 int형 변수를 받겠다고 선언뒤 for문으로 원소들을 하나씩불러와 합치는 방식으로 하였습니다.
<br/>
<br/>

```python
def addall(li):
    return sum([x for x in li])

print(addall([1]))
print(addall([1,2,3,4,5]))
```

#### addall에서 sum을 사용하여 sum안에 list comprehension으로 list를 만들어주어 합계가 리턴되도록 만들었습니다.
<br/>

2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

```python
def addallodd(li):
    result = 0
    for i in li:
        if i%2 == 1:
            result += i
    return result

print(addallodd([1]))
print(addallodd([1,2,3,4,5,6,7,8,9]))
```

#### addallodd에 list인자를 받은뒤 int형을 받기위해 result=0을 선언해준뒤 for문으로 원소를 하나씩받아 그것이 홀수이면 결과값에 더해주는 식으로 합을 구하였습니다.
<br/>

```python
def addallodd(li):
    return sum([x for x in li if x%2==1])

print(addallodd([1]))
print(addallodd([1,2,3,4,5,6,7,8,9]))
```

#### addallodd에 list인자를 받은뒤 sum에 list comprehension을 이용하여 인자로 받은 list에서 홀수만을 받는 리스트를 주어 홀수만의 합을 갖는 결과값을 반환하도록하였습니다.

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
key = list(d.keys())
key.sort(reverse=True)
for i in key:
    a = (i, d[i])
    print(a)
```

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
d_reverse = dict(map(reversed, d.items()))
value = list(d.values())
value.sort()
for i in value:
    a = (i, d_reverse[i])
    print(a)
```

### 8.다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.

length(X)를 다음과 같이 정의한다 : 10진수로 표현된 X의 자릿수

Strange라는 숫자를 아래과 같이 정의한다.

음이 아닌 정수 중 자릿수가 1인 모든 숫자는 Strange이다.

length(X) > 1 인 숫자 X 는 다음 2가지를 만족하면 Strange이다.

X는 length(X)로 정확히 나누어 떨어진다.

X / length(X) 가 strnage이다.

```python
def isStrange(num):
    length = 0
    tmp = num
    while True:
        length += 1
        tmp = tmp//10
        if tmp == 0:
            break;
    if length == 1:
        return True
    else:
        if num%length == 0:
            return isStrange(num/length) # or False or isStrange(매개변수 숫자)
        else:
            return False
print("1. num = 0                     =>    ",isStrange(0))
print("2. num = 56                    =>    ",isStrange(56))
print("3. num = 100                   =>    ",isStrange(100))
print("4. num = 7882602506492313600   =>    ",isStrange(7882602506492313600))
print("5. num = 6012302549522098651   =>    ",isStrange(6012302549522098651))
```

#### 먼저 while문을 통해 num의 값을 받은것을 나눠주면서 자릿수를 구한다.
#### 만약 length가 1이면 True를 반환 해주고, 아니라면 num이 length로 나눠떨어지는지 확인해준다.
#### 만약 나누어 떨어진다면 return isStrange(num/length)를 해주어 num/length한값을 isStrange보내주어 결과값을 반환하도록 재귀함수를 호출한다.
#### 나누어 떨어지지 않는다면 False를 출력해준다.

### 9.다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.
1번 그룹, 2번 그룹, ... n번 그룹으로 번호가 매겨진 총 n개의 그룹이 존재하고, ith 그룹에는 임의의 팀원 수 ai 명이 있다.
하나의 버스 정류장이 있고, 버스는 오직 한 대만 운영한다. 버스는 버스 정류장에 도착하고 모든 그룹은 버스를 통해 여행하길 원한다.
하지만, 같은 그룹에 속한 팀원들은 분리되기를 원치 않는다. 그래서 그룹의 모든 팀원이 탈 수 있을 때 에만 그 그룹은 버스에 탑승한다.
또한, 이들은 그룹의 번호순으로 여행을 하길 원한다. 예를 들어, 3번 그룹은 1번 그룹과 2번 그룹이 여행을 한 뒤에 버스를 탈 수 있거나, 1번 그룹과 2번 그룹이 이미 버스에 탑승한 상태에 있을 때 동시에 버스를 탈 수 있다.
마지막으로, 버스는 빈 좌석이 없을 때 에만 여행을 떠난다.

이 문제에서 각 그룹의 팀원 수가 리스트로 주어질 때, 가능한 모든 버스의 사이즈 x(x명의 사람들이 동시에 탈 수 있음)를 반환하는 함수 solve(a) 를 작성하시오.

[주의] 반드시 계산 복잡도를 최대한 낮추세요. 아래 예시 3번을 실행했을 때 결과가 바로 나타날 수 있도록!!

```python
def check(a, num):
    i=0
    temp = num
    while i!=len(a):
        temp -= a[i]
        if temp == 0 and i!=len(a)-1:
            temp = num
        elif temp <0:
            return False
        i+=1
    if temp == 0:
        return True
            
def solve(a):
    result =[]
    lim = sum(a)
    i=0
    number = 0
    for j in range(0,len(a)):
        number += a[j]
        if number>lim/2:
            break;
        if lim%number == 0 and check(a, number):
            result.append(number)
    result.append(lim)
    return result

a = [1, 2, 1, 1, 1, 2, 1, 3] 
print("1. a = [1, 2, 1, 1, 1, 2, 1, 3]    =>   ", solve(a))
a = [1, 1, 1, 1]
print("2. a = [1, 1, 1, 1]                =>   ", solve(a))
a = [9999] * 10**5
print("3. a = [9999] * 10**5              =>   ", solve(a))
```

#### solve에서는 결과값을 출력하는 리스트 result와 무조건 들어가는 값(리스트의 모든수를 더한값) lim을 선언해주었고, 앞에서 하나씩 더해가며 조건문을 통과한것만이 버스의 사이즈가 되도록하였다.
#### 결과의 공통점을 보면 버스의 사이즈는 리스트의 모두(sum)를 더한값에서 나누어 떨어진다는 것을 알 수 있었다. 또한 나누어 떨어진다고 하더라도 그 값들이 순차적으로 뺐을때 모든 좌석을 다채우는지를 확인해야 하였고 check함수를 만들었다.
#### check함수는 받은 리스트와 값을 이용해서 값에서 list의 앞에서부터 끝까지 하나씩 빼보면서 초과하는 값이 나타나면 false를 리턴하고 한번도 빠짐없이 0으로 끝나며 리스트의 끝까지 갔을때도 좌석을 다채우면 True를 리턴함으로써 조건을 만족하는지 확인하도록 하였다.
#### 위의 조건을 만족하게되면 결과리스트에 append해주고 마지막에 result를 리턴해줌으로써 결과값을 출력하게 된다.

### [Incremental Project 문제]
이전 Assignment 2의 마지막 문제는 웹 URL로 지정된 웹페이지를 문자열로 가져와 모든 HTML 태그를 제외한 순수 텍스트를 얻어내고 그 안에 존재하는 단어를 추출하고 단어들의 총 개수를 출력하는 프로그램을 작성하는 것이었다. 이번에는 그 마지막 숙제를 그대로 확장하여 웹 URL로 지정된 웹페이지 내 순수 텍스트 안에 존재하는 각 단어들에 대해 다음 요구사항 대로 출력하는 프로그램을 작성하시오.

```python
import requests
import string
req = requests.get('https://www.naver.com/')

source = req.text
check = 0
check_slash =0
total = ""
for i in range(len(source)):
    if (source[i:i+7] == '<script' or source[i:i+6] == '<style') and check_slash == 0:
            check_slash = 1
    elif (source[i:i+9] == '</script>' or source[i:i+8] == '</style>') and check_slash == 1:
            check_slash = 0

    if check == 0:
        if source[i] =="<":
            check =1
        else:
            if check_slash ==0:
                total += source[i]
    else:
        if source[i] == ">":
            check =0
li = total.split()
result = []
for i in li:
    for punctuation in string.punctuation:
        i = i.replace(punctuation, "")
    result.append(i)
result = list(filter(None, result))
word_dic = {}
for i in result:
    if i in word_dic:
        word_dic[i] += 1
    else:
        word_dic[i]=1
print("단어수 : ",len(result))
print(word_dic)
```
